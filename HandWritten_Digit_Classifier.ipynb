```python
import pandas as pd
import numpy as np
from tensorflow import keras
import matplotlib.pyplot as plt
```

```python
(x1,y1),(x2,y2)=keras.datasets.mnist.load_data()
```

```python
x1[0].shape
```

```python
x1.shape
```

```python
x1=x1/250
x2=x2/250 #scaling
```

```python
plt.matshow(x1[4])
```

```python
x3=x1.reshape(len(x1),28*28)
x4=x2.reshape(len(x2),28*28)
x4.shape
```

```python
x1[0] #2D array
```

```python
x3[0] # 1D array
```

```python
reg=keras.Sequential([keras.layers.Dense(10,input_shape=(784,),activation='sigmoid')])
```

```python
reg.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
reg.fit(x3,y1,epochs=5)
```

```python
reg.evaluate(x4,y2)
```

```python
reg.predict(x4)
```

```python
plt.matshow(x1[2])
```

```python
y=reg.predict(x3)
y[2]
```
